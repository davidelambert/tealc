# Environment

```python
import pandas as pd
from numpy import power
from statsmodels.api import OLS
from plotnine import *
```

```python
def scatterfit(data, x, y, plot_title,
               point_color='black', line_color='black'):
    """Quick scatterplot with OLS fit line using plotnine.ggplot
    
    data -- a Pandas dataframe
    x, y -- [string] column names from df
    plot_title -- [string]
    point_color, line_color -- [string]
    Returns: plotnine.ggplot object
    """
    
    p = ggplot(data, aes(x, y)) +\
        stat_smooth(method='ols', color=line_color,
                    fill=line_color, alpha=0.3) +\
        geom_point(color=point_color, size=2) +\
        labs(title=plot_title)

    return p
```

```python
def short_table(fitted_model):
    """Abbreviated single-variable OLS results table"""
    
    coef = round(fitted_model.params['gauge_sq'], ndigits=4)
    se = round(fitted_model.bse['gauge_sq'], ndigits=4)
    t = round(fitted_model.tvalues['gauge_sq'], ndigits=4)
    p = round(fitted_model.pvalues['gauge_sq'], ndigits=4)
    r2 = round(fitted_model.rsquared, ndigits=4)

    print('Coef'.rjust(14) + 'SE'.rjust(14) + 't'.rjust(14) +
          'Pr>|t|'.rjust(14) + 'R-squared'.rjust(14))
    print(str(coef).rjust(14) + str(se).rjust(14) + str(t).rjust(14) +
          str(p).rjust(14) + str(r2).rjust(14))
```

# Plain Steel

## Manufacturer G

```python
g_ps = pd.read_csv('../charts/ps.csv')
g_ps['gauge_sq'] = power(g_ps['gauge'], 2)

g_ps_plot = scatterfit(g_ps, x='gauge_sq', y='unit_weight',
                         plot_title='Mfr. G plain steel',
                         point_color='dimgrey', line_color='grey')
g_ps_plot.draw();
```

```python
g_ps_fit = OLS(g_ps['unit_weight'], g_ps['gauge_sq']).fit()
short_table(g_ps_fit)
```

## Large-gauge nickel-plated steel

In Manufacturer G's published unit weight charts, there appear to be abrupt discontinuities
in unit weights for the 0.056" and 0.058" gauges, then again for gauges 0.060"
and above. Though no other Manufacturer G electric sets are as comprehensive as their nickel-plated steel offering,
the unit weights for these larger gauges appear to be consistent with other Manufacturer G
string types, so I don't think the discontinuities are a printing mistake.
Perhaps they are an artifact of construction, such as abrupt jumps in the ratio
of core gauge to wrap gauge.

Due to the discontinuities, I estimate a separate unit weight model for larger
gauges. Though the 56 &amp; 58 gauges don't seem to fit well in either
the small or large gauge classes, I include them in the large gauge model
because their inclusion has a very small effect on gauge_sq
coefficient/standard error (.2058/.0056 included, .2059/.0066 excluded) or on
R-squared (.9935, .9928).

The final tension estimation program should include control flow to calculate
tensions for nickel-plated steel strings above gauge 0.054" separately.

```python
lg_nps = pd.read_csv('../charts/nps_large.csv')
lg_nps = lg_nps[lg_nps['gauge'] > 0.054]
lg_nps['gauge_sq'] = power(lg_nps['gauge'], 2)

lg_nps_plot = scatterfit(lg_nps, 'gauge_sq', 'unit_weight',
                         plot_title='Mfr. G large-gauge nickel-plated steel',
                         point_color='steelblue', line_color='cornflowerblue')
lg_nps_plot.draw();
```

```python
lg_nps_fit = OLS(lg_nps['unit_weight'], lg_nps['gauge_sq']).fit()
short_table(lg_nps_fit)
```
